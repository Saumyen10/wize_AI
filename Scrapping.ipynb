```python
#library
import requests
from bs4 import BeautifulSoup
```

```python
#selecting an author to scrap data
url = "https://medium.com/@mrdbourke"

data = requests.get(url)
soup = BeautifulSoup(data.text, 'html.parser')
```

```python
print(soup)
```

```python
# Function to remove unwanted phrases from cleaned content
def remove_unwanted_phrases(text):
    text = str(text)
    # List of unwanted phrases
    unwanted_phrases = ["Daniel Bourke", "inTowards Data Science"]

    # Loop through the unwanted phrases and replace them with an empty string
    for phrase in unwanted_phrases:
        text = text.replace(phrase, '')

    # Return the cleaned text
    return text.strip()
```

```python
import re
from bs4 import BeautifulSoup

def clean_text(text):

    text = text.lower()                     # Convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)     # Remove punctuation
    text = re.sub(r'\d+', '', text)
    text = text.strip()
    return text

# Parse the HTML content
soup = BeautifulSoup(data.text, 'html.parser')

# Scraping the article data
for article in soup.find_all('article'):
    title = article.find('h2').text
    content = article.find('div').text

    # Extracting date from its text
    date = article.find('span', class_='')

    # Remove the title from content
    cleaned_content = content.replace(title, '')

    # Regex to remove date-like patterns
    cleaned_content = re.sub(r'(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?) ?\d{1,2}', '', cleaned_content)

    # Strip any extra spaces or leftover punctuation
    cleaned_content = cleaned_content.strip()

    # Pass the cleaned content to the remove_unwanted_phrases function
    cleaned_content = remove_unwanted_phrases(cleaned_content)

    # Now use the cleaned content in the clean_text function
    final_cleaned_content = clean_text(cleaned_content)

    # Output the results
    print(f"Title: {title}\nContent: {final_cleaned_content}\nDate: {date.text}\n")
```
